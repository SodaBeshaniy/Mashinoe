## ***Типизация***

```
df.Age = df.Age.astype("float")

df.ChestPainType = df.ChestPainType.astype("category")

df.Cholesterol	= df.Cholesterol.astype("int")

df.dtypes
```

# ***Фильтрация***

```
df[df["Age"] > 50]
```

```
df[df["Cholesterol"] > 200]
```
