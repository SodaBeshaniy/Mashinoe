##Типизация

```python
df.astype('object')

df['Birth Rate'] = df['Birth Rate'].astype('float64')
df['Calling Code'] = df['Calling Code'].astype('float64')
df['Fertility Rate'] = df['Fertility Rate'].astype('float64')
df['Infant mortality'] = df['Infant mortality'].astype('float64')
df['Life expectancy'] = df['Life expectancy'].astype('float64')
df['Maternal mortality ratio'] = df['Maternal mortality ratio'].astype('float64')
df['Physicians per thousand'] = df['Physicians per thousand'].astype('float64')
df['Latitude'] = df['Latitude'].astype('float64')
df['Longitude'] = df['Longitude'].astype('float64')

df['Country'] = df['Country'].astype('category')
df['Abbreviation'] = df['Abbreviation'].astype('category')
df['Capital/Major City'] = df['Capital/Major City'].astype('category')
df['Currency-Code'] = df['Currency-Code'].astype('category')
df['Largest city'] = df['Largest city'].astype('category')
df['Official language'] = df['Official language'].astype('category')

df.info()
```
